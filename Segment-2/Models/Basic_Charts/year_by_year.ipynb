```python
import os
import pandas as pd
import numpy as np
```

```python
# Import Polaris dataset downloaded...
eda_df = pd.read_csv(os.path.join("Potential_Data", "global_dataset_14_Apr_2020.csv"), header=0)

year_counts_df = pd.DataFrame(eda_df['yearOfRegistration'].value_counts()).reset_index()
year_counts_df.columns = ['year', 'case_count']
year_counts_df.sort_values(by=['year'], inplace=True)
year_counts_df
```

```python
import matplotlib.pyplot as plt
x = year_counts_df['year'].astype('str')
y = year_counts_df['case_count']
plt.figure(figsize=(12,8))
plt.bar(x, y)
plt.title("Case Count by Year")
plt.xlabel("Year", fontsize='large')
plt.xticks(rotation = 45)
plt.ylabel("Case Count")
plt.show()
```

```python
country_counts = pd.DataFrame(eda_df.groupby(['citizenship'])['yearOfRegistration'].value_counts())
country_counts.rename(columns = {'yearOfRegistration': 'yearCount'}, inplace=True)
country_counts
```
