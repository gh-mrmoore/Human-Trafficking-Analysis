# Feature Engineering

## Means of Control
### Table 1: Means of Control Groupings
| Grouped Category | Data Category | Data Code |
| ----- | ----- | ----- |
| Financial | meansOfControlDebtBondage  | 1 |
| Financial | meansOfControlTakesEarnings | 1 |
| Financial | meansOfControlRestrictsFinancialAccess | 1 |
| Threats | meansOfControlThreats | 2 |
| Threats | meansOfControlPsychologicalAbuse | 2 |
| Threats | meansOfControlFalsePromises | 2 |
| Threats | meansOfControlUsesChildren | 2 |
| Threats | meansOfControlThreatOfLawEnforcement | 2 |
| Survival | meansOfControlRestrictsMedicalCare | 3 |
| Survival | meansOfControlWithholdsNecessities | 3 |
| Survival | meansOfControlWithholdsDocuments | 3 |
| Physical | meansOfControlPhysicalAbuse | 4 |
| Physical | meansOfControlSexualAbuse | 4 |
| Physical | meansOfControlRestrictsMovement | 4 |
| Physical | meansOfControlExcessiveWorkingHours | 4 |
| Other | meansOfControlOther | 5 |
| Other | meansOfControlNotSpecified | 5 |
| Other | meansOfControlPsychoactiveSubstances | 5 |


```python
# Categorize by type of control per table above
control_means_list = []
for row in eda_df.iterrows():
# for row in eda_subset_df.iterrows():
    # Categorize by arbitrary grouping of 'Financial'
    if str(row[1]["meansOfControlDebtBondage"]) == "1" or str(row[1]["meansOfControlTakesEarnings"]) == "1" or str(row[1]["meansOfControlRestrictsFinancialAccess"]) == "1":
        control_means_list.append('Financial')
    # Categorize by arbitrary grouping of 'Threats'
    elif str(row[1]["meansOfControlThreats"]) == "1" or str(row[1]["meansOfControlPsychologicalAbuse"]) == "1" or str(row[1]["meansOfControlFalsePromises"]) == "1" or str(row[1]["meansOfControlUsesChildren"]) == "1" or str(row[1]["meansOfControlThreatOfLawEnforcement"]) == "1":
        control_means_list.append('Threats')
    # Categorize by arbitrary grouping of 'Survival'
    elif str(row[1]["meansOfControlRestrictsMedicalCare"]) == "1" or str(row[1]["meansOfControlWithholdsNecessities"]) == "1" or str(row[1]["meansOfControlWithholdsDocuments"]) == "1":
        control_means_list.append('Survival')
    # Categorize by arbitrary grouping of 'Physical'
    elif str(row[1]["meansOfControlPhysicalAbuse"]) == "1" or str(row[1]["meansOfControlSexualAbuse"]) == "1" or str(row[1]["meansOfControlRestrictsMovement"]) == "1" or str(row[1]["meansOfControlExcessiveWorkingHours"]) == "1":
        control_means_list.append('Physical')
    # Categorize by arbitrary grouping of 'Other'
    elif str(row[1]["meansOfControlNotSpecified"]) == "1" or str(row[1]["meansOfControlPsychoactiveSubstances"]) == "1" or str(row[1]["meansOfControlOther"]) == "1":
        control_means_list.append('Other')
    # Insert null value if none of the criteria above are met...
    else:
        # print("None")
        control_means_list.append(np.nan)
```

```python
# Add a column with our arbitrary category groupings back to the main DataFrame
eda_df["ControlCategory"] = control_means_list
eda_df["ControlCategory"].value_counts()
```

```python
# Duplicate DataFrame and drop columns that we no longer need...
eda1_df = eda_df
eda1_df.drop(["meansOfControlDebtBondage", "meansOfControlTakesEarnings", "meansOfControlRestrictsFinancialAccess"], axis=1, inplace=True)
eda1_df.drop(["meansOfControlThreats", "meansOfControlPsychologicalAbuse", "meansOfControlFalsePromises", "meansOfControlUsesChildren", "meansOfControlThreatOfLawEnforcement"], axis=1, inplace=True)
eda1_df.drop(["meansOfControlRestrictsMedicalCare", "meansOfControlWithholdsNecessities", "meansOfControlWithholdsDocuments"], axis=1, inplace=True)
eda1_df.drop(["meansOfControlPhysicalAbuse", "meansOfControlSexualAbuse", "meansOfControlRestrictsMovement", "meansOfControlExcessiveWorkingHours"], axis=1, inplace=True)
eda1_df.drop(["meansOfControlNotSpecified", "meansOfControlPsychoactiveSubstances", "meansOfControlOther"], axis=1, inplace=True)
eda1_df.drop(["meansOfControlConcatenated"], axis=1, inplace=True)
eda1_df.head()
```

## Recruiter Relationship

### Table 2: Recruiter Relationships Categorized
| Original Group | Categorized Group | Data Code |
| ----- | ----- | ----- |
| Not Specified | Not Specified | 1 |
| Other | Other | 2 |
| Not Specified; Other | Other | 2 |
| Friend/Acquaintance | Friend/Acquaintance | 3 |
| Friend/Acquaintance; Not Specified; Other | Friend/Acquaintance | 3 |
| Friend/Acquaintance; Not Specified | Friend/Acquaintance | 3 |
| Friend/Acquaintance; Intimate Partner; Not Specified | Friend/Acquaintance | 3 |
| Friend/Acquaintance; Other | Friend/Acquaintance | 3 |
| Friend/Acquaintance; Intimate Partner | Friend/Acquaintance | 3 |
| Friend/Acquaintance; Intimate Partner; Other | Friend/Acquaintance | 3 |
| Family/Relative | Family/Relative | 4 |
| Family/Relative; Other | Family/Relative | 4 |
| Family/Relative; Not Specified | Family/Relative | 4 |
| Family/Relative; Intimate Partner | Family/Relative | 4 |
| Family/Relative; Friend/Acquaintance | Family/Relative | 4 |
| Family/Relative; Friend/Acquaintance; Other | Family/Relative | 4 |
| Family/Relative; Intimate Partner; Other | Family/Relative | 4 |
| Intimate Partner; Other | Intimate Partner | 5 |
| Intimate Partner; Not Specified | Intimate Partner | 5 |
| Intimate Partner | Intimate Partner | 5 |


```python
# Re-categorize RecruiterRelationship
relation_categories = []
for row in eda1_df.iterrows():
    if row[1]["RecruiterRelationship"] == "Not Specified":
        relation_categories.append('Not Specified')
    elif row[1]["RecruiterRelationship"] == "Not Specified; Other" or row[1]["RecruiterRelationship"] == "Other":
        relation_categories.append('Other')
    elif row[1]["RecruiterRelationship"] == "Friend/Acquaintance" or \
        row[1]["RecruiterRelationship"] == "Friend/Acquaintance; Not Specified; Other" or \
            row[1]["RecruiterRelationship"] == "Friend/Acquaintance; Not Specified" or \
                row[1]["RecruiterRelationship"] == "Friend/Acquaintance; Intimate Partner; Not Specified" or \
                    row[1]["RecruiterRelationship"] == "Friend/Acquaintance; Other" or \
                        row[1]["RecruiterRelationship"] == "Friend/Acquaintance; Intimate Partner" or \
                            row[1]["RecruiterRelationship"] == "Friend/Acquaintance; Intimate Partner; Other":
        relation_categories.append('Friend/Acquaintance')
    elif row[1]["RecruiterRelationship"] == "Family/Relative" or \
        row[1]["RecruiterRelationship"] == "Family/Relative; Other" or \
            row[1]["RecruiterRelationship"] == "Family/Relative; Not Specified" or \
                row[1]["RecruiterRelationship"] == "Family/Relative; Intimate Partner" or \
                    row[1]["RecruiterRelationship"] == "Family/Relative; Friend/Acquaintance" or \
                        row[1]["RecruiterRelationship"] == "Family/Relative; Friend/Acquaintance; Other" or \
                            row[1]["RecruiterRelationship"] == "Family/Relative; Intimate Partner; Other":
        relation_categories.append('Family/Relative')
    elif row[1]["RecruiterRelationship"] == "Intimate Partner" or \
        row[1]["RecruiterRelationship"] == "Intimate Partner; Other" or \
            row[1]["RecruiterRelationship"] == "Intimate Partner; Not Specified":
        relation_categories.append('Intimate Partner')
    else:
        relation_categories.append(np.nan)
```

```python
# Add our revised category listing to our DataFrame
eda1_df["RecruiterCategory"] = relation_categories
eda1_df["RecruiterCategory"].value_counts()
```

```python
# Duplicate DataFrame and drop unnecessary recruiter columns
eda2_df = eda1_df
eda2_df.drop(["RecruiterRelationship", "recruiterRelationIntimatePartner", "recruiterRelationFriend", "recruiterRelationFamily", "recruiterRelationOther", "recruiterRelationUnknown"], axis=1, inplace=True)
eda2_df.head()
```

## Type of Exploit

### Table 3: Exploit Types Categorized
| Original Data | Data Code |
| ----- | ----- |
| Sexual exploitation | 1 |
| Forced labour | 2 |
| Other | 3 |
| Slavery and similar practices | 4 |
| Forced marriage | 5 |
| -99 (Unknown) | NaN |

```python
# Re-categorize typeOfExploitConcatenated
exploit_categories = []
for row in eda2_df.iterrows():
    if str(row[1]["typeOfExploitConcatenated"]) == "Sexual exploitation":
        exploit_categories.append('Sexual Exploitation')
    elif str(row[1]["typeOfExploitConcatenated"]) == "Forced labour":
        exploit_categories.append('Forced Labor')
    elif str(row[1]["typeOfExploitConcatenated"]) == "Other":
        exploit_categories.append('Other')
    elif str(row[1]["typeOfExploitConcatenated"]) == "Slavery and similar practices":
        exploit_categories.append('Slavery and Similar')
    elif str(row[1]["typeOfExploitConcatenated"]) == "Forced marriage":
        exploit_categories.append('Forced Marriage')
    elif str(row[1]["typeOfExploitConcatenated"]) == "-99":
        exploit_categories.append(np.nan)
    else:
        exploit_categories.append(3)
```

```python
# Add re-categorized data to DataFrame
eda2_df["ExploitType"] = exploit_categories

# Duplicate our database
eda3_df = eda2_df

# Drop columns
eda3_df.drop(["typeOfExploitConcatenated", \
                         "isSexualExploit", \
                         "isOtherExploit", \
                         "isSexAndLabour", \
                         "isForcedMarriage", \
                         "isForcedMilitary", \
                         "isOrganRemoval", \
                         "isSlaveryAndPractices"], axis=1, inplace=True)
eda3_df.head()
```

```python
# Unite the 'typeOf' columns to a single column
labor_types = []
for row in eda3_df.iterrows():
    if row[1]["typeOfLabourAgriculture"] == 1:
        labor_types.append('Agriculture')
    elif row[1]["typeOfLabourAquafarming"] == 1:
        labor_types.append('Aquafarming')
    elif row[1]["typeOfLabourBegging"] == 1:
        labor_types.append('Begging')
    elif row[1]["typeOfLabourConstruction"] == 1:
        labor_types.append('Construction')
    elif row[1]["typeOfLabourDomesticWork"] == 1:
        labor_types.append('Domestic Work')
    elif row[1]["typeOfLabourHospitality"] == 1:
        labor_types.append('Hopitality')
    elif row[1]["typeOfLabourIllicitActivities"] == 1:
        labor_types.append('Illicit Activities')
    elif row[1]["typeOfLabourManufacturing"] == 1:
        labor_types.append('Manufacturing')
    elif row[1]["typeOfLabourMiningOrDrilling"] == 1:
        labor_types.append('Mining or Drilling')
    elif row[1]["typeOfLabourPeddling"] == 1:
        labor_types.append('Peddling')
    elif row[1]["typeOfLabourTransportation"] == 1:
        labor_types.append('Transportation')
    elif row[1]["typeOfLabourOther"] == 1 or row[1]["typeOfLabourNotSpecified"] == 1:
        labor_types.append('Other')
    else:
        labor_types.append(12)    
```

```python
# Add labor categories to our DataFrame
eda3_df["Labor_Type"] = labor_types
eda3_df["Labor_Type"].value_counts()
```
